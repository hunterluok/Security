```python
def nchosey(n, y):
    temp = math.factorial(n)
    temp_1 = math.factorial(y)
    temp_2 = math.factorial(n - y)
    return temp / (temp_1 * temp_2)
nchosey(10, 3)
```

```python
def get_e(y, n=10, a=7, b=5):
    cons1 = nchosey(n, y)
    cons2 = gamma(a + b) / (gamma(a) * gamma(b))
    cons3 = gamma(a + y) * gamma(b + n - y) / (gamma(a + b + n))
    return cons1 * cons2 * cons3
```

```python
get_e(14, n=20, a=1, b=1)
get_e(14, n=20, a=50, b=50)
get_e(14, n=20, a=5, b=1)
```
