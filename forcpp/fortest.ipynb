```python
import pandas as pd

from sklearn.datasets import load_iris
import matplotlib.pyplot  as plt
%matplotlib inline

data = load_iris()['data']
target = load_iris()['target']
```

```python
resu = pd.read_csv("resultt.txt", header=None, sep=" ")
resu.columns = ['indexs', 'dist']
pred = resu.indexs.values

def plot(label=pred, first=0, second=1):
    if first == second:
        raise ValueError(" equal is wrong")
    plt.scatter(data[label==1, first], data[label==1, second])
    plt.scatter(data[label==2, first], data[label==2, second])
    plt.scatter(data[label==0, first], data[label==0, second])
    plt.show()
```

```python
plot(label=target)
```

```python
plot(second=1, first=2)
```

```python
plot(second=0, first=2)
```

```python
plot(second=0, first=1)
```
